```python
import pyfolio as pf
import pandas as pd
import numpy as np
from yahoo_finance import Share
import matplotlib
import datetime
from datetime import datetime
from sklearn.linear_model import LinearRegression
```

```python
gs = pd.DataFrame(Share('GS').get_historical('2010-05-02', '2015-12-12'))
gs.Date = [datetime.strptime(gs.Date.iloc[i], '%Y-%m-%d') for i in gs.index]
gs.index = gs.Date
gs = gs.iloc[::-1]
gs.Close = [float(x) for x in gs.Close]
```

```python
plot(gs.Close)
```

```python
def get_dailyret(ser):
    return (ser.shift(-1) - ser)/ser
```

```python
def get_smooth_val(ser, lookback):
    ###does not include today so that residual is relevant for return
    ls = []
    bs = []
    lr = LinearRegression()
    for i in range(len(ser)):
        a = ser[i-lookback:i].tolist()
        #print(a)
        if len(a) == 0:
            ls.append(np.nan)
            bs.append(np.nan)
        else:
            a = lr.fit(np.array(range(lookback)).reshape(lookback,1),np.array(a).reshape((lookback,1)))
            ls.append(float(a.predict(lookback)[0]))
            bs.append(float(a.coef_))
    return ls, bs
```

```python
m,n = get_smooth_val(gs.Close, 50)
gs['smooth'], gs['trend'] = m,n#pd.Series(m), pd.Series(n)
```

```python
gs['resid'] = gs.Close-gs['smooth']
```

```python
plot(gs[ 'resid'])
```

```python
gs['dailyrets'] = get_dailyret(gs.Close)
```
